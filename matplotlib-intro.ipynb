## Matplotlib workflow example

```python
# 0.import and get jupyter ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x = [1,3,5,7]
y = [11,33,55,77]

# 2.setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3.plot data
ax.plot(x,y);

# 4.customize plot
ax.set(title = "simple sample",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5.save and show figure
fig.savefig("images/sampleplotfig.png")
```

pyplot vs matplotlib oo method:
pyplot can be used when need quick visualization and oo method when something more advanced needed.

```python
over50 = heartdis[heartdis['age']>50]
over50.head()
```

```python
len(over50)
```

pyplot method

```python
over50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target', figsize=(10,6));
#c here denotes colour and try with and wihtout fig size
```

oo method

```python
fig , ax = plt.subplots(figsize = (10,6))
over50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target', ax = ax);
#ax = ax tells us we are using oo method
ax.set_xlim([45,100])
```

```python
fig , ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(x = over50['age'], y = over50['chol'], c = over50['target']);
ax.set(title = 'heart disease & cholestrol levels',
      xlabel = 'age', ylabel = 'cholestrol');
ax.legend(*scatter.legend_elements(), title = 'target');
```

```python
over50.target.values
```

```python
#add horizontal line
fig , ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(x = over50['age'], y = over50['chol'], c = over50['target']);
ax.set(title = 'heart disease & cholestrol levels',
      xlabel = 'age', ylabel = 'cholestrol');
ax.legend(*scatter.legend_elements(), title = 'target');
ax.axhline(over50['chol'].mean(),
          linestyle = '--');
```

```python
over50.head()
```

add subplots to dataframes : OO method use

```python
#subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
#add data to ax0
scatter = ax0.scatter(x=over50["age"], y=over50["chol"], c=over50["target"])
#customize ax0
ax0.set(title='heart disease & cholestrol levels', xlabel='age', ylabel='cholestrol')
#add legend to ax0
ax0.legend(*scatter.legend_elements(), title='target');
#add a meanline
ax0.axhline(y=over50['chol'].mean(), linestyle='--');

#add data to ax1
scatter = ax1.scatter(x=over50['age'], y=over50['thalach'], c=over50['target'])
#customize ax1
ax1.set(title='heart disease & max heart rate', xlabel='age', ylabel='heart rate(MAX)');
#add legend to ax1
ax1.legend(*scatter.legend_elements(), title='target')
#add a meanline
ax1.axhline(y=over50['thalach'].mean(), linestyle='--');

#add a title to entire figure
fig.suptitle("heart disease", fontsize=16, fontweight='bold');
```

```python
#set the style
plt.style.use('seaborn-v0_8-whitegrid')
#OO method from the scratch
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter = ax.scatter(x=over50['age'],
                    y=over50['chol'],
                    c=over50['target'])
#customize the plot
ax.set(title='heart diseases and cholestrol levels',
      xlabel='age', ylabel='cholestrol')
#add a legend
ax.legend(*scatter.legend_elements(), title='target');
#add a horizontal line
ax.axhline(over50['chol'].mean(),linestyle='--');
```

```python
#in the output you dont like the colour so you add cmap parameter that changes the colour scheme
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over50['age'],
                    y=over50['chol'],
                    c=over50['target'],
                    cmap='winter') #try cmap as summer and plasma too and check out matlotlib colour tutorial for more such styles.
ax.set(title='heart diseases and cholestrol levels',
      xlabel='age', ylabel='cholestrol')
ax.legend(*scatter.legend_elements(), title='target');
ax.axhline(over50['chol'].mean(),linestyle='--');
```

customizing x and y limitations

```python
#code for plot with no modifications, the below plot has gone through modifications
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
scatter = ax0.scatter(x=over50["age"], y=over50["chol"], c=over50["target"])
ax0.set(title='heart disease & cholestrol levels', xlabel='age', ylabel='cholestrol')
ax0.legend(*scatter.legend_elements(), title='target');
ax0.axhline(y=over50['chol'].mean(), linestyle='--');
scatter = ax1.scatter(x=over50['age'], y=over50['thalach'], c=over50['target'])
ax1.set(title='heart disease & max heart rate', xlabel='age', ylabel='heart rate(MAX)');
ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(y=over50['thalach'].mean(), linestyle='--');
fig.suptitle("heart disease", fontsize=16, fontweight='bold');
```

```python
#subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
scatter = ax0.scatter(x=over50["age"],
                      y=over50["chol"],
                      c=over50["target"],
                      cmap='winter')
ax0.set(title='heart disease & cholestrol levels', xlabel='age', ylabel='cholestrol')
ax0.set_xlim([50,80]) #for cleaner edges
ax0.legend(*scatter.legend_elements(), title='target');
ax0.axhline(y=over50['chol'].mean(), linestyle='--');
scatter = ax1.scatter(x=over50['age'],
                      y=over50['thalach'],
                      c=over50['target'],
                      cmap='plasma')
ax1.set(title='heart disease & max heart rate', xlabel='age', ylabel='heart rate(MAX)');
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(y=over50['thalach'].mean(), linestyle='--');
fig.suptitle("heart disease", fontsize=16, fontweight='bold');
```
